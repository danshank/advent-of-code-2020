```python
import re

file = open("day-7-input", "r")
lines = file.read().splitlines()

m = {}
        
for line in lines:
    c = re.match(r"([\w|\s]+) bags contain", line)
    s = re.findall(r"((\d+|no) ([\w|\s]+) bags?,?)+", line)
    bag = c.groups(1)[0]
    for cs in s:
        if cs[2] == "other":
            continue
        elif cs[2] in m:
            m[cs[2]].add(bag)
        else:
            x = set()
            x.add(bag)
            m[cs[2]] = x

sg = "shiny gold"
x = set()

def search(graph, found, x):
    found.add(x)
    if x in graph:
        nextSearch = graph[x]
        for y in nextSearch:
            if y not in found:
                search(graph, found, y)
                
search(m, x, sg)

print(len(x) - 1)
```

```python
import re

file = open("day-7-input", "r")
lines = file.read().splitlines()

m = {}
        
for line in lines:
    c = re.match(r"([\w|\s]+) bags contain", line)
    s = re.findall(r"((\d+|no) ([\w|\s]+) bags?,?)+", line)
    bag = c.groups(1)[0]
    reg = {}
    for cs in s:
        if cs[2] == "other":
            continue
        else:
            reg[cs[2]] = cs[1]
    m[bag] = reg

sg = "shiny gold"

def bagsInBag(graph, bag):
    count = 0
    for b in graph[bag]:
        print(b)
        count = count + int(graph[bag][b]) * (1 + bagsInBag(graph, b))
    return count
                
print(bagsInBag(m, sg))
```
